```python
#program 3
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} â†’ Pass"
    else:
        return f"Average: {average:.2f} â†’ Fail"

# Example
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print(result)
```

```python
#program 4
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(km, peak_hour)
        print(f"ðŸš– Total Fare: â‚¹{total_fare:.2f}")
        break

    except:
        print("âŒ Invalid input. Please try again.")
```

```python
#program 5
def check_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% â†’ Eligible"
    else:
        return f"Attendance: {percentage:.2f}% â†’ Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_eligibility(attendance))
```

```python
#program 6
def check_password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "âŒ Weak Password: Minimum length should be 8."

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "âœ… Strong Password"
    else:
        return "âŒ Weak Password: Must contain at least one digit and one special character (@#$)."

# Example
pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
